```python
### Transformación de los datos en la columna Date
```

```python
df['Date']= pd.to_datetime(df['Date'], utc=True, errors='coerce')
df['Year']=df['Date'].dt.year
df['Year'] = pd.to_numeric(df['Year'], errors='coerce').fillna(0).astype(int)
df.head()
```

### Reemplazo de valores en la columna Price

```python
df["Price"] = df["Price"].fillna("0") #Rellenamos los valores nulos con 0 y transformamos los valores a float
df['Price'] = pd.to_numeric(df['Price'], errors = 'coerce')
df["Price"] = df["Price"].astype(float)
df.head()
```

### Obtenemos la columna Country

```python
df['Country'] = df['Location'].apply(lambda x : x.split(',')[-1].strip())
df.head()
```

## Paso 3: Consultas

### ¿Qué organización lanzó más misiones espaciales en un año determinado?

```python
spf_year = 2010  # Cambia esto por el año que desees analizar

# Agrupa por año y organización, cuenta los registros
conteo_año = df.groupby(['Year', 'Organisation']).size().reset_index(name='conteo')

# Filtra por el año específico
conteo_año_filtrado = conteo_año[conteo_año['Year'] == spf_year]

# Encuentra la organización con el máximo conteo
organizacion_max = conteo_año_filtrado.loc[conteo_año_filtrado['conteo'].idxmax()]

# Imprime el resultado
print(f"La organización con el mayor número de lanzamientos en el año {spf_year} es:")
print(organizacion_max)
```

```python
# Ordenar el DataFrame por 'conteo' en orden descendente
conteo_año_filtrado = conteo_año_filtrado.sort_values(by='conteo', ascending=True)

# Crear un gráfico de barras horizontales
plt.figure(figsize=(10, 6))  # Tamaño de la figura (ancho, alto)
bars = plt.barh(conteo_año_filtrado['Organisation'], conteo_año_filtrado['conteo'], color='red')

# Configurar el color de fondo del área del gráfico
plt.gca().set_facecolor('black')  # Color de fondo del área de trazado

# Añadir títulos y etiquetas
plt.xlabel('Número de Lanzamientos')
plt.ylabel('Organización')
plt.title(f'Número de Lanzamientos por Organización en el Año {spf_year}')

# Añadir los valores en las barras
for bar in bars:
    plt.text(
        bar.get_width() + 0.2,  # Posición x (un poco más a la derecha del final de la barra)
        bar.get_y() + bar.get_height() / 2,  # Posición y (centro vertical de la barra)
        f'{int(bar.get_width())}',  # Texto del valor (convertido a entero)
        va='center',  # Alineación vertical (centrado)
        ha='left',    # Alineación horizontal (a la izquierda)
        color='white' # Color del texto
    )

# Mostrar el gráfico
plt.show()
```

## ¿Qué meses son los más populares para los lanzamientos? 

```python
df['Month']=df['Date'].dt.month
df['Month'] = pd.to_numeric(df['Month'], errors='coerce').fillna(0).astype(int)
df.head()
```

```python
plt.bar(df.groupby('Month')['Date'].count().index,
         df.groupby('Month')['Date'].count().values,
       color = "red")
plt.gca().set_facecolor('black')

# Obtener los valores de las barras
values = df.groupby('Month')['Date'].count().values
months = df.groupby('Month')['Date'].count().index

# Añadir los valores sobre las barras
for i, value in enumerate(values):
    plt.text(months[i], value + 5, str(value), ha='center', color='white')

# Etiquetas de los ejes (opcional)
plt.xlabel('Mes')
plt.ylabel('Conteo de Fechas')

# Mostrar el gráfico
plt.show()
```
